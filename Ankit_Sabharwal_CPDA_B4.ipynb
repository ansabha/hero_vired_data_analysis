#### Mechant_state, zip and errors have NaN.
- **FOR zip:** I first drop the `NaN` values and then convert the remaining values to strings. This is necessary because numeric columns containing `NaN` are treated as floats by default. If we directly convert them to strings, values may appear as something like `'nan.0'`, which would then be treated as a valid value.

  The process is:

  1. Drop the `NaN` values.
  2. Convert the remaining values to integers.
  3. Convert the integers to strings.
  4. Reindex the data to bring back the rows that were dropped because of `NaN`.
  5. Fill those missing rows with empty strings.
  6. Add leading zeroes where required.

-  **For merchant_state:** We have change the state code to ISO code for the ones which were not there. For example, chnaged China to CH. Also fgigured out that there were about 8008 entries which were NaN. These entries were related to ONLINE transaction. For missing values, these transactions are related to ONLIne, thus have put ONLINE over there.

- **For use_chip:** Changed it to include only Swipe, Chip and online

```python
import pandas as pd
import numpy as np


df_total_limits = cards_df.groupby('client_id')['credit_limit'].sum().reset_index()
df_total_limits.columns = ['client_id', 'total_credit_limit']


df_user_risk = pd.merge(users_df, df_total_limits, left_on='id', right_on='client_id', how='left')
df_user_risk['total_credit_limit'] = df_user_risk['total_credit_limit'].fillna(0)
df_user_risk['credit_to_income_ratio'] = df_user_risk['total_credit_limit'] / df_user_risk['yearly_income']
df_user_risk['total_debt_to_yearly_income_ratio']=df_user_risk['total_debt'] / df_user_risk['yearly_income']

conditions = [
    (df_user_risk['credit_to_income_ratio'] <= 0.50) & (df_user_risk['total_debt_to_yearly_income_ratio'] <= 2),
    (((df_user_risk['credit_to_income_ratio'] >= 0.50) & (df_user_risk['total_debt_to_yearly_income_ratio'] <= 2)) |
     ((df_user_risk['credit_to_income_ratio'] <= 0.50) & (df_user_risk['total_debt_to_yearly_income_ratio'] >= 2))),
    (df_user_risk['credit_to_income_ratio'] > 0.50) & (df_user_risk['total_debt_to_yearly_income_ratio'] >=2)
]
risk_labels = ['Safe', 'Moderate Risk', 'High Risk']

df_user_risk['risk_tier'] = np.select(conditions, risk_labels, default='No Income Record')
```

```python
df_user_risk['risk_tier'].value_counts()
```

```python
risk_counts = df_user_risk['risk_tier'].value_counts()
plt.figure(figsize=(8, 8))

colors = {
    'High Risk': '#f0ad4e',       # Soft Orange
    'Moderate Risk': '#f0de4e',   # Soft Yellow
    'Safe': '#5cb85c'             # Soft Green
}

plot_colors = [colors.get(tier, '#999999') for tier in risk_counts.index]

plt.pie(
    risk_counts, 
    labels=risk_counts.index, 
    autopct='%1.1f%%',          # Formats percentages to one decimal place (e.g., '24.0%')
    startangle=140,              # Rotates the start point for a cleaner alignment
    colors=plot_colors, 
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}  # Adds crisp gaps between segments
)

# 4. Final layout customizations
plt.title('User Financial Risk Profile Distribution', fontsize=15, fontweight='bold', pad=20)
plt.tight_layout()

# Show the plot
plt.show()
```
